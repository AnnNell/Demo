```python
#importing libraries

import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#read the image data

img1= cv2.imread ('plane.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img1, cmap='gray')
plt.show()
```

```python
#Convert an BGR to RGB
img2_BGR= cv2.imread ('plane.jpg', cv2.IMREAD_COLOR)
img2_RGB= cv2.cvtColor(img2_BGR, cv2.COLOR_BGR2RGB)
plt.imshow(img2_RGB)
plt.show()
```

```python
#To save a file
cv2.imwrite("plane_new.jpg", img1)
```

```python
#Resize the image 50x50
img_50x50= cv2.resize(img1, (50,50))
plt.imshow(img_50x50, cmap='gray')
plt.show()
```

```python
#crop the image
img_cropped= img_50x50[:,:20]
plt.imshow(img_cropped, cmap='gray')
plt.show()
```

```python
#enhance image
image_enhanced = cv2.equalizeHist(img1)
plt.imshow(image_enhanced, cmap='gray')
plt.show()
```

```python
image_bgr = cv2.imread('plane_256x256.jpg')
image_hsv = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2HSV)
plt.imshow(image_hsv)
plt.show()
```

```python
lower_blue = np.array([50,100,50])
upper_blue = np.array([130,255,255])
```

```python
mask = cv2.inRange(image_hsv, lower_blue, upper_blue)
plt.imshow(mask, cmap='gray')
plt.show()
```

```python
#Detecting Edges

# Load image as grayscale
image_gray = cv2.imread("plane_256x256.jpg", cv2.IMREAD_GRAYSCALE)

# Calculate median intensity
median_intensity = np.median(image_gray)

# Set thresholds to be one standard deviation above and below median intensity
lower_threshold = int(max(0, (1.0 - 0.33) * median_intensity))
upper_threshold = int(min(255, (1.0 + 0.33) * median_intensity))

# Apply canny edge detector
image_canny = cv2.Canny(image_gray, lower_threshold, upper_threshold)

# Show image
plt.imshow(image_canny, cmap="gray"), plt.axis("off")
plt.show()
```
