### Impute categorical values

```python
df_new=df
```

```python
def impute_nan_most_frequent_category(df_new,ColName):
  
    df_new[ColName + "_Imputed"] = df_new[ColName]
    df_new[ColName + "_Imputed"].fillna(df_new[ColName].mode()[0],inplace=True)
```

```python
for Columns in ['Cabin','Embarked']:
    impute_nan_most_frequent_category(df_new,Columns)
    
df_new.head()
```
